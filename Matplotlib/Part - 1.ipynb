````python
# from a pandas dataframe
batsman = pd.read_csv('/content/sharma-kohli.csv')
batsman

plt.plot(batsman['index'],batsman['V Kohli'])
````

````python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
````

````python
# labels & tittle
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit Vs Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Score')
````

````python
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6')

plt.title('Rohit Vs Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Score')
````

````python
# linestyle(dashed,dashdot,solid etc.)
# linewidth(as your choice like 1,3,5,4 etc.)
plt.plot(batsman['index'],batsman['V Kohli'],color='yellow',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dashdot',linewidth=2)

plt.title('Rohit Vs Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Score')
````

````python
# marker (D,+,.,>,<,o)
# markersize (as your choice like 5,7,8,10 etc.)
plt.plot(batsman['index'],batsman['V Kohli'],color='yellow',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dashdot',linewidth=2,marker='o',markersize=7)

plt.title('Rohit Vs Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Score')
````

````python
# legend -> location
plt.plot(batsman['index'],batsman['V Kohli'],color='yellow',linestyle='solid',linewidth=3,marker='D',markersize=8,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dashdot',linewidth=2,marker='o',markersize=7,label='Rohit')

plt.title('Rohit Vs Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Score')

plt.legend()
````

### Google Colab Scatter Plot Note

1. **Chote Data:**
   - Use `plt.scatter()`
   - Color aur formatting options available.

2. **Bade Data:**
   - Use `plt.plot()`
   - Formatting options nahi hote.

#### Example:
```python
# Chote Data
plt.scatter(x_small, y_small)

# Bade Data
plt.plot(x_large, y_large)
```

# Bar chart

- Bivariate Analysis
- numerical vs categorical
- Use case - Aggregate analysis of groups

````python
childern = [10,20,40,10,30]
color = ['red','blue','green','yellow','pink']
````

````python
# simple bar chart
plt.bar(color,childern,color=['red','blue','green','yellow','pink'])
plt.xlabel('Color',color='red')
plt.ylabel('Children',color='red')
plt.title('Childern Favourite Color')
````

````python
# horizontal Bar Chart using plt.barh
# ye tab use karte hai jab categorical value jyada ho..

plt.barh(color,childern,color='yellow')
plt.xlabel('Color',color='red')
plt.ylabel('Children',color='red')
plt.title('Childern Favourite Color')
````

````python
# color and label
df = pd.read_csv('/content/batsman_season_record.csv')
df
````

````python
# Season 2015, 2016, 2017 combine bar chat

plt.bar(np.arange(df.shape[0]) - 0.2 ,df['2015'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]) ,df['2016'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]) + 0.2 ,df['2017'],width=0.2,color='blue')

# this is used for replace the name with batsman
plt.xticks(np.arange(df.shape[0]),df['batsman'])

plt.xlabel('Batsman',color='red')
plt.ylabel('Runs',color='red')
plt.title('Runs in 2015 season By Batsman')


plt.xticks(rotation=90)
# isse ham thoda rotation change kar dete hai
plt.show()
````

# Pie Chart

- Univariate/Bivariate Analysis
- Categorical Numerical
- Use case - To find contibution on a standard scale

````python
# sample data
data = [23,45,100,20,49]

plt.pie(data)
````

````python
# plt.show use krne se additional information hide ho jata hai
plt.pie(data)
plt.show()
````

````python
data = [23,45,100,20,49]
subject = ['eng','science','maths','sst','hindi']

plt.pie(data,labels=subject)
plt.show()
````

````python
# dataset
df = pd.read_csv('/content/gayle-175.csv')
df
````

````python
plt.pie(df['batsman_runs'],labels=df['batsman'])
plt.show()
````

### percentage and colors

````python
# autopct use krne se percentage show hota hai pi chart me

plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
````

````python
# explode
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0])
plt.show()
````

````python
# shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0],shadow=True)
plt.show()
````
