# Annotations
- use krne se pata laga sakte hai ki plot me kya indicate or kisko kar rahi hai
- Example ek scatter plot hai cricket ka usme runs hai to ham usme identify kar sakte hai ki kon batsman hai

```python
batters = pd.read_csv('/content/batter.csv')
batters.head()
```

```python
batters.shape
```

```python
sample_df = batters.head(100).sample(25,random_state = 5)
sample_df
```

```python
plt.figure(figsize=(18,10))

plt.scatter(sample_df['avg'],sample_df['strike_rate'])
plt.xlabel('Average')
plt.ylabel('Strike Rate')

# ye for loop lagane se ham pata laga sakte hai kon battersman kitne run or avg se khel raha tha (name) ke sath
for i in range (sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
# Ab ham run ke according points ko bada karenge jisse pata chale ki kon sa battersman kitne strike
# or avg se khel raha hai or jo battersman run jyada marega usko bade points se indicate karenge

plt.figure(figsize=(18,10))

# ,s=sample_df['runs'] maine add kiya hai (s for size)
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
plt.xlabel('Average')
plt.ylabel('Strike Rate')


for i in range (sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

#### Different Between Scatter Plot vs Subplots

- Scatter Plot

```python
plt.figure(figsize=(6,3))
plt.scatter(batters['avg'],batters['strike_rate'])
plt.title('Something')
plt.xlabel('Average')
plt.ylabel('Strike Rate')

plt.show()
```

- Subplots

```python
# Upar wale plot ko print karana
fig, ax = plt.subplots(figsize=(10,5))

ax.scatter(batters['avg'],batters['strike_rate'],color='red',marker='+')

# isme title dene ke liye use this ->>
ax.set_title('Something')

# Thik isi tarah xlabel or ylabel bhi
ax.set_xlabel('Average')
ax.set_ylabel('Strike Rate')

fig.show()
```

```python
# batter dataset
```

```python
# iske use se ham multiple plot banate hai
# 2 Row or 1 Column karenge to 2 plot banega
# 2 row or 2 column karenge to 4 plots banenge

fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True)

ax[0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[1].scatter(batters['avg'],batters['runs'],color='green',marker='^')

ax[0].set_title('Average Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')

ax[1].set_title('Average Vs Runs')
ax[1].set_xlabel('Average')
ax[1].set_ylabel('Runs')
```

## 4 Graphs creation with the help of Subplots

```python
# jab ham 2*2 ka plot banate hai matlab jab 4 plot create krte hai tab hame
# 2-d array milta hai yani ki upar mai ham 2 plot create kiye the to access
# karne ke liye ax[0], ax[1] karte hai

# usi tarah ham agar 2*2 ka plot banate hai tab ham access krne ke liye
# ax[0,0], ax[0,1], ax[1,0], ax[1,1] karte hai
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0,1].scatter(batters['avg'],batters['runs'],color='yellow')
ax[1,0].hist(batters['avg'],color='green')
ax[1,1].hist(batters['runs'],color='black')


ax[0,0].set_title('Average Vs Strike Rate')
ax[0,1].set_title('Average Vs Runs')
ax[1,0].set_title('Average')
ax[1,1].set_title('Runs')
```

```python
fig = plt.figure()

# (2,1,1) -->> 3 rows 1 column or 1 graph
ax1 = fig.add_subplot(3,1,1)
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

# (2,1,2) -->> 3 rows 1 column or 2 graph
ax2 = fig.add_subplot(3,1,2)
ax2 = plt.scatter(batters['avg'],batters['runs'],color='green')

# (2,1,2) -->> 3 rows 1 column or 3 graph
ax3 = fig.add_subplot(3,1,3)
ax3 = plt.hist(batters['runs'],color='yellow')
```

```python
fig = plt.figure()

# (2,1,1) -->> 2 rows 2 column or 1 graph
ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

# (2,1,2) -->> 2 rows 2 column or 2 graph
ax2 = fig.add_subplot(2,2,2)
ax2 = plt.scatter(batters['avg'],batters['runs'],color='green')

# (2,1,2) -->> 2 rows 2 column or 3 graph
ax3 = fig.add_subplot(2,2,3)
ax3 = plt.hist(batters['runs'],color='yellow')
```

# 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
yy.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,6))
ax = plt.subplot(projection='3d')

ax.plot_surface(xx,yy,z,cmap='viridis')
```

```python
fig = plt.figure(figsize=(12,6))
ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
j = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,j,cmap='viridis')
fig.colorbar(p)
```

```python
j = np.sin(xx) + np.log(yy)
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,j,cmap='viridis')
fig.colorbar(p)

plt.show()
```

# Contour Plots

```python
fig = plt.figure(figsize=(12,6))

ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

# Heatmap

```python
delivery = pd.read_csv('/content/IPL_Ball_by_Ball_2008_2022.csv')
delivery.sample()
```

```python
# find the most sixes in overs
temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
```

```python
# Create Pivot table in the based on how many 6 sixes in the over
grid = temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
plt.figure(figsize=(20,10))
# heatmap create krne ke liye plt.imshow use kare
plt.imshow(grid)

# ye 1,2,3,4...20 over by over show ho raha hai
plt.yticks(delivery['overs'].unique(), list(range(1,21)))

# ye 1,2,3,4,5,6 ball dikhata hai over ke liye
plt.xticks(np.arange(0,6), list(range(1,7)))

plt.colorbar()
```
