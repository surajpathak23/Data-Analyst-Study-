## Seaborn Roadmap
### Types of function
- Figure Level (isme legend axes se bahar hota hai)
  - is plot me agar plot ka size manage krna ho to `height` & `aspect` Use kare.
  - For Example -> sns.displot(tips,x='total_bill',kind='kde',hue='sex',fill=True,`height=7,aspect=2`)
- Axis Level (isme legend plot ke anadar hota hai)
  - is plot ka agar figure size manage karna ho to matplotlib se bhi kar sakte hai
  - for Example -> `plt.figure(figsize=(15,10))`

# Scatter plot

```python
# Dataset
tips = sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(tips,x='total_bill',y='tip')
```

### `scatterplot` -> (hue, style, size)

```python
# `hue` -> ise use krne se aap alag alag category ko different color se represent kara sakte hai
# `hue` -> isme ham jiska category divide karna ho to `hue` ke anadar likhte hai

# `style` -> ise use karne se time me bhi category devide karke dikhata hai

# `size` -> ise use krne se marker ka size bada dete hai according to the number

sns.scatterplot(tips,x='total_bill',y='tip',hue='sex',style='time',size='size')
```

```python
# relplot -> figure level -> square shape

# Same graph with the help of 'relplot'

sns.relplot(tips,x='total_bill',y='tip',hue='sex',style='time',size='size')
```

# 2. Distribution Plots
- `Used for univariate analysis` (Sirf ek column par Analysis like-Salary)
- `Used to find out the distribution` (kaha mujhe acha result mil sakta hai)
- `Range of the observation` (max-min ke bich me result milta hai)
- `Central Tendency` (min kya hai data ka median.? kaha par center hai data ka)
- `is the data bimodal ?` (bimodal ka matlb hai multple picks hai ky..?? data me)
- `Are there Outliers ?` (kya.?? Outliers hai data me)
## plots under distribution plot
- histplot
- kdeplot
- rugplot


### `Note âœˆ`
- figure level -> displot
- axes level -> histplot , kdeplot , rugplot
- `displot` use krke ham all plot bana sakte hai but isme bhi `kind='hist' 'kde'` karke bana sakte hai
- Agar displot use krke plot create krte hai or kind use nhi karenge to bydefault plot `histplot` banega


```python
# plotting univariate histogram
sns.histplot(tips,x='total_bill')
```

### Using `displot`

```python
sns.displot(tips,x='total_bill',kind='hist')
```

#### `bins` parameter
- Bins ke help se ham number of plot ka column devide karte hai ki mujhe itne bins ka plot bana kar do
- Upar me `bins` Bydefault `14` hai
- Niche me bins 20 hai using `bins=20`

```python
sns.displot(tips,x='total_bill',kind='hist',bins=20)
```

- Apply `histplot` in category column like- day(Thur, Fri, Sat, Sun)

```python
sns.displot(tips,x='day',kind='hist')
```

### `hue` Parameter
- alag-alag category me divide karke dikhana
- like sex me `Male` & `Female` ko

```python
sns.displot(tips,x='total_bill',kind='hist',hue='sex')
```

### `element` -> step
- ise ham `hue` ke sath use krte hai
- iske help se thoda ache se dikhata hai categorical label
- like upar me Male & Female ko ache se visulation krne ke liye `element='size'` use kiye hai

```python
sns.displot(tips,x='total_bill',kind='hist',hue='sex',element='step')
```

## `Kde Plot`
- `Kde` -> Kernel Density Estimation

```python
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernal, producing a continuous density estimate
# y axes par density
sns.kdeplot(tips,x='total_bill')
```

### `hue` -> Use for show the category in plot
- like -sex(Male & Female)

#### `fill` -> Use for fill the Area of the Category
- like - Male & Female me transparate color fill ho gya `fill=True` use karne se

```python
# height -> use for manage the hight of plot
# aspect -> height ke double size width ho jata hai
# Note -> Agar aspect nhi denge to automatic width manage kar deta hai
sns.displot(tips,x='total_bill',kind='kde',hue='sex',fill=True,height=7,aspect=2)
```

## Rugplot

```python
# Plot marginal distributions by drawing ticks along the x and y axes
# This function is intended is complement other plots by showing the location of individual observations in an unobtrusive way

sns.kdeplot(tips,x='total_bill')
sns.rugplot(tips,x='total_bill')
```

#### Bivariate histogram

```python
# A bivariate histogram bins the data within rectangles that tile the plot
# and then shows the count of observations within each rectangle with the fill color

# sns.histplot(tips,x='total_bill',y='tip')
# Or
sns.displot(tips,x='total_bill',y='tip',kind='hist')
```

#### Bivariate Kdeplot

```python
# a bivariate KDE plot smooth the (x,y) observations with a 2D Gaussian

sns.kdeplot(tips,x='total_bill',y='tip')
```

# Matrix Plot
- Heatmap
- Clustermap

### `Heatmap`

```python
gap
```

```python
# Plot rectangular data as a color-encoded matrix
temp_df = gap.pivot(index='country',columns='year',values='lifeExp')
temp_df
```

#### Create Heatmap of temp_df
- `Heatmap` -> axes level function

```python
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
# Europe Continent ke according Heatmap
temp_df = gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')

plt.figure(figsize=(15,10))
sns.heatmap(temp_df)
```

####

#### `annot=True` -> Use karne se `Heatmap me numbers` bhi show hone lagte hai
#### `linewidth` -> Iske use se ham `Heatmap` me `box ke bich ka gap` de sakte hai
#### `cmap` -> Isko use krne se ham `Heatmap colorful` bana sakte hai

```python
plt.figure(figsize=(15,10))
sns.heatmap(temp_df,annot=True,linewidth=0.5,cmap='viridis')
```
