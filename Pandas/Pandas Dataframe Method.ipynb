# Creating Dataframe
---
#### `pd.DataFrame()`
- `pd.DataFrame([...], columns=[...])` → नया DataFrame banata hai.
- Data ko **rows** aur **columns** ke form me store karta hai.
- Example me `Name, IQ, Marks, Package` columns ke sath table create hua.

---

##### `.set_index('col_name')` Method
- `.set_index('Name')` → given column ko **index** bana deta hai.

# More Important Functions

#### `.value_counts()` Method

##### On **Series**
- `series.value_counts()` → har unique value ka **count** batata hai.
- Useful jab frequency distribution dekhna ho.
- Example: `a.value_counts()` → batayega ki Series `a` me kaunsi value kitni bar aayi hai.

##### On **DataFrame**
- `dataframe.value_counts()` → puri row kitni bar repeat hui hai wo count karta hai.
- Example: `marks.value_counts()` → batayega ki kaunsi row (all columns same) kitni bar repeat hui hai.

### Find the last match played by virat kohli in Delhi

```python
# Team1 or Team2 ke players ko add kar liye
ipl['all_players'] = ipl['Team1Players'] + ipl['Team2Players']
ipl.head(1)
```

```python
def did_kohli_play(players_list):
  return 'V Kohli' in players_list
```

```python
ipl['did_kohli_play'] = ipl['all_players'].apply(did_kohli_play)
# column created with name of 'did_kohli_play
ipl.sample(1)
```

```python
# Kohli play in delhi all matches
ipl[(ipl['City'] == 'Delhi') & (ipl['did_kohli_play'] == True)].sample(2)
```

```python
# last match play kohli in Delhi
ipl[(ipl['City'] == 'Delhi') & (ipl['did_kohli_play'] ==True)].drop_duplicates(subset=['City','did_kohli_play'],keep='first')
```

```python
# DataFrame in apply function
points_df = pd.DataFrame({
    '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
    '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
})
points_df
```

```python
def euclidean(row):
  pt_A = row['1st point']
  pt_B = row['2nd point']

  return ((pt_A[0] - pt_B[1])**2 + (pt_A[1] - pt_B[1])**2)**0.5
```

```python
points_df['distance'] = points_df.apply(euclidean,axis=1)
points_df
```
