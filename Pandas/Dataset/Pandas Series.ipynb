# ðŸ“˜ Pandas Series â€“ Cheat Sheet

## ðŸ”¹ Basics
- `type(s)` â†’ Series ka type check
- `dir(s)` â†’ Available functions & attributes
- `len(s)` â†’ Length of series
- `list(s)` â†’ Convert Series to list
- `sorted(s)` â†’ Sorted list of values

---

## ðŸ”¹ Attributes
- `s.index` â†’ Index (row labels)
- `s.values` â†’ Values as array
- `s.dtype` â†’ Data type of Series
- `s.shape` â†’ Shape (rows,)
- `s.size` â†’ Total elements
- `s.ndim` â†’ Dimensions (1D)
- `s.is_unique` â†’ Values unique hain ya nahi

---

## ðŸ”¹ Indexing & Slicing
- `s[0]` â†’ Single value
- `s[1:5]` â†’ Slice values
- `s.loc[]` â†’ Label-based indexing
- `s.iloc[]` â†’ Position-based indexing
- `s[[0,2,3]]` â†’ Fancy indexing (multiple values)

---

## ðŸ”¹ Head & Tail
- `s.head()` â†’ First 5 values
- `s.tail()` â†’ Last 5 values

---

## ðŸ”¹ Type Conversion
- `s.astype('category')` â†’ Change data type

---

## ðŸ”¹ Membership
- `'value' in s.values` â†’ Check value exist karta hai ya nahi

---

## ðŸ”¹ Operators
- `100-s` â†’ Arithmetic operation on all values
- `s>=50` â†’ Relational check (True/False)

---

## ðŸ”¹ Boolean Indexing
- `s[s>50]` â†’ Filter values
- `(s==0).sum()` â†’ Count zero values

---

## ðŸ”¹ Sorting & Counting
- `s.sort_values()` â†’ Sort values
- `s.value_counts()` â†’ Frequency count
- `s.count()` â†’ Non-missing values count

---

## ðŸ”¹ Aggregation
- `s.max()` â†’ Maximum
- `s.min()` â†’ Minimum
- `s.mean()` â†’ Average
- `s.sum()` â†’ Total sum
- `s.product()` â†’ Product of values
- `s.describe()` â†’ Summary stats

---

## ðŸ”¹ Uniqueness
- `s.unique()` â†’ Unique values
- `s.nunique()` â†’ Count of unique values

---

## ðŸ”¹ Missing Data
- `s.isnull()` â†’ NaN check
- `s.notnull()` â†’ Non-NaN check
- `s.fillna(0)` â†’ Replace NaN
- `s.dropna()` â†’ Remove NaN

### Series with Custom `Index`

#Series Method

### `.head()` Method
  - `.head()` â†’ top 5 rows return karta hai by default.
  - Useful jab hum large dataset ke sirf starting records dekhna chahte hain.

````python
bolly.head()
````

````python
kohli.head(7)
````

### `.tail()` Method  
- `.tail()` â†’ last 5 rows return karta hai by default.  
- Large dataset me jab hume **ending records** dekhne ho tab useful hai.  
- Example: `subs.tail()` â†’ `subs` DataFrame ki last 5 rows show karega.  


````python
subs.tail()
````

````python
bolly.tail(9)
````

### `.sample()` Method  
- `.sample()` â†’ dataset se **random row/rows** return karta hai.  
- By default 1 random row deta hai.  
- Useful jab hume dataset ka **random observation** dekhna ho.  
- Example: `kohli.sample()` â†’ `kohli` DataFrame se 1 random row show karega.  


````python
kohli.sample()
````

````python
bolly.sample(6)
````

### `.value_counts()` Method  
- `.value_counts()` â†’ column me unique values ka **frequency count** return karta hai.  
- Data ko summarize karne aur **most common values** dekhne ke liye useful hai.  
- Example: `bolly['lead'].value_counts()` â†’ `lead` column me har actor/actress kitni baar aaya hai, uska count dikhata hai.  
- ye `descending order` me value deta hai


````python
bolly['lead'].value_counts()
````

### `.sort_values()` Method  
- `.sort_values()` â†’ DataFrame/Series ko kisi **column ke values ke basis par sort** karta hai.  
- By default ascending order me sort hota hai.  
- Example: `kohli.sort_values('runs')` â†’ `kohli` DataFrame ko `runs` column ke values ke hisaab se ascending order me arrange karega.  
- Agar descending order chahiye ho to:  
  ```python
  kohli.sort_values('runs', ascending=False)


````python
kohli.sort_values('runs')
````

````python
kohli.sort_values('runs',ascending=False)
````

### Top 5 Runs using `.sort_values()` + `.head()`  
- `kohli.sort_values('runs', ascending=False)` â†’ `runs` column ke basis par data ko **descending order** me sort karega.  
- `.head()` â†’ top 5 rows return karega.  
- Combined: `kohli.sort_values('runs', ascending=False).head()` â†’ Kohli ke IPL data me **top 5 highest runs wali rows** show karega.  


````python
kohli.sort_values('runs',ascending=False).head()
````

### Extract Top Run Value Only  
- `kohli.sort_values('runs', ascending=False)` â†’ `runs` ke basis par data ko descending order me arrange karega.  
- `.head(1)` â†’ top 1 row return karega (highest run wali).  
- `.values[0]` â†’ us row ka **sirf numpy array format me values** extract karega.  
- Example: `kohli.sort_values('runs', ascending=False).head(1).values[0]` â†’ Kohli ke sabse highest run wali row ke saare column values return karega (numpy array ke form me).  


````python
kohli.sort_values('runs',ascending=False).head(1).values[0]
````

### `.sort_values()` on "Subscribers gained" (Descending Order)  
- `.sort_values('Subscribers gained', ascending=False)` â†’ Data ko `Subscribers gained` column ke values ke basis par **descending order** me sort karega.  
- Isse hume pata chalega ki **kis channel ne sabse zyada subscribers gain kiye hain**.  
- Example: `subs.sort_values('Subscribers gained', ascending=False)` â†’ `subs` DataFrame ko highest â†’ lowest subscribers gained ke order me arrange karega.  


````python
subs.sort_values('Subscribers gained',ascending=False)
````

### `.sort_values()` with `inplace=True`  
- `inplace=True` â†’ changes ko **original DataFrame me directly apply** kar deta hai, naya copy return nahi karta.  


````python
kohli.sort_values('runs',inplace=True)
kohli
````

### `.sort_values()` on "movie" Column with `inplace=True`  
- `.sort_values('movie', ascending=False, inplace=True)` â†’ `movie` column ke basis par data ko **descending alphabetical order** me sort karega.  
- `inplace=True` â†’ sorting ka effect **directly original DataFrame** par apply hoga.  

````python
bolly.sort_values('movie',ascending=False, inplace=True)
bolly
````

### Python Built-in Functions with Pandas Objects  

1. **`len(object)`**  
   - DataFrame/Series me kitni rows hain, uska count deta hai.  
   - Example: `len(subs)` â†’ `subs` DataFrame ki total rows ka number.  

2. **`type(object)`**  
   - Object ka data type batata hai.  
   - Example: `type(subs)` â†’ `<class 'pandas.core.frame.DataFrame'>`.  

3. **`dir(object)`**

   - Example: `dir(subs)` â†’ Pandas DataFrame ke methods (jaise `head`, `tail`, `sort_values`, etc.).

   - Example: `sorted(subs)` â†’ `subs` ke columns sorted list me.

   - Columns ke labels me se **maximum (last in alphabetical order)** return karta hai.  
   - Example: `max(subs)` â†’ alphabetically last column name.  

6. **`min(object)`**  
   - Columns ke labels me se **minimum (first in alphabetical order)** return karta hai.  
   - Example: `min(subs)` â†’ alphabetically first column name.

````python
# ye batata hai ki kitne bar 50 se jyada runs mara hai
fifties = kohli[(kohli['runs'] >= 50)]
fifties.size

# ye adjust value nhi batayega ki kohli 50's se jyada run kitni bar mara hai
````

````python
# Count of 50s (between 50 and 99)
fifties = kohli[(kohli['runs'] >= 50) & (kohli['runs'] < 100)].shape[0]

# Count of 100s (100 or more)
hundreds = kohli[kohli['runs'] >= 100].shape[0]

print("50s:", fifties)
print("100s:", hundreds)
````
