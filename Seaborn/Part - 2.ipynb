## Boxplot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (minimum, first quartile [Q1], median, third quartile [Q3] and maximum). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

image.png

### `Boxplot

````python
sns.boxplot(tips,x='day',y='total_bill',hue='sex')
````

# Categorical Estimate Plot
## `barplot`
- Ek `categorical` & Ek `numerical` column ke sath koi `aggrigate value` calculate karna
- `ci=None` -> Use karne se `Error-bar` hat jata hai

````python
# ye average bata rah hai
sns.barplot(tips,x='sex',y='total_bill')
````

When there are multiple observations in each category, It also uses bootstrapping to compute a confidence interval around the estimate, which in plotted using error bars

````python
# isme hum std nikal rahe hai
sns.barplot(tips,x='sex',y='total_bill',hue='smoker',estimator=np.std,ci=None)
````

````python
# median ke liye
sns.barplot(tips,x='sex',y='total_bill',hue='smoker',estimator=np.median)
````

````python
# min ke liye
# show in the plot -> male smoker me minimum total_bill aaya hai 7$ ka

sns.barplot(tips,x='sex',y='total_bill',hue='smoker',estimator=np.min)
````

## `countplot`
A special case for the bar plot is when you want to show the number of observations in each category rather than computing a statistic for a second variable. This is similar to a histogram over a categorical, rather than quantitative, variable

````python
# It's represent the count of Male & Female
sns.countplot(tips,x='sex')
````

````python
# sns.countplot(tips,x='sex',hue='day')
# or
sns.catplot(tips,x='sex',hue='day',kind='count')
````

## Regression Plots
- regplot (Axes level Function) -> `hue` not support
- lmplot (Figure level Function) -> `hue` support

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

- > Note :-
  - `hue` parameter is not available in `regplot`
  - `lmplot` use krne se `same` rseult milega `regplot` ki tarah But, isme `kind use nhi hota hai`

````python
sns.regplot(tips,x='total_bill',y='tip')
````

````python
sns.lmplot(tips,x='total_bill',y='tip',hue='sex')
````

### Plotting Pairwise Relationship (PairGrid Vs Pairplot)

````python
iris.sample()
````

````python
# sns.pairplot(iris)
sns.pairplot(iris,hue='species')
````

### `PairGrid`

````python
g = sns.PairGrid(iris,hue='species')

# g.map
g.map(sns.scatterplot)
````

````python
# map_diag -> map_offdiag

g = sns.PairGrid(iris,hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.kdeplot)
````

````python
# map_diag -> map_upper -> map_lower
g = sns.PairGrid(iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
````

#### `var`
- vars=['sepal_width','petal_width']
- iske use se ham in dono ka plot banayega

````python
g = sns.PairGrid(iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
````
